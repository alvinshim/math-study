## 명제
명제 : 참, 거짓을 명확히 판별할 수 있는 문장 또는 식  
ex) 2는 소수이다.  
  
조건 p를 만족하는 진리집합 P  
조건 q를 만족하는 진리집합 Q  
p 또는 q : $P \cup Q$  
p 그리고 q : $P \cap Q$  
  
U = {x | x는 10 이하의 자연수}  
p : x는 짝수이다.  
$\to P=\{2,4,6,8,10\}$  
q = x는 소수이다.  
$\to Q=\{2,3,5,7\}$  
  
### 명제 또는 조건의 부정  
2는 소수이다(참) : 2는 소수가 아니다(거짓)  
4는 소수이다(거짓) : 4는 소수가 아니다(참)  
  
p의 부정 : ~p (not p)  
~P : $P^{c}$  
  
p 또는 q의 부정 : ~p 그리고 ~q  
$\to(P \cup Q)^{c}=P^{c}\cap Q^{c}$  
  
p 그리고 q의 부정 : ~p 또는 ~ q  
$\to(P \cap Q)^{c}=P^{c}\cup Q^{c}$  
  
#### $p\to q$의 참/거짓  
p : x는 4의 배수이다  
q : x는 2의 배수이다  
$p\to q$ : x가 4의 배수이면 x는 2의 배수  
$P=\{4,8,12,16,\dots\}$  
$Q=\{2,4,6,8,\dots\}$  
$P \subset Q$  
$P\to Q$ : 참  
  
### '모든', '어떤' 이 포함된 명제의 참/거짓  
모든 정수는 양수이다. (거짓)  
위 명제의 부정 : 어떤 정수는 양수가 아니다.  
  
'모든'이 포함된 명제가 참이 되려면,  
$P=U$  
'어떤'이 포함된 명제가 참이 되려면,  
$P\neq\emptyset$  
  
### 명제의 역과 대우  

| $p\to q$ | $\underleftrightarrow{역}$ | $q\to p$ |
|--|--|--|
| 대우$\updownarrow$ | -- | 대우$\updownarrow$ |
|~$q\to$ ~$p$| $\underleftrightarrow{역}$ |~$p\to$ ~$q$| 

원 명제와 대우 명제는 참과 거짓을 함께한다.  
원 명제가 참/거짓이라고 해도 역 명제가 반드시 참/거짓이라고 할 수 없다.  
  
$p \implies q$  
~$p \implies$ ~$r$  
~$q \implies$ ~$p$  
$r \implies p$  
~$q \implies$ ~$p \implies$~$r$ :   ~$q \implies$ ~$r$  
$r \implies p \implies q$ : $r \implies q$  
  
귀류법 : 결론을 부정하여 모순이 생김을 보는 것
