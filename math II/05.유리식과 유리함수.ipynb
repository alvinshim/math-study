## 유리식  
### 유리식 기초  
A, B 다항식  
$\frac{A}{B} \gets \frac{다항식}{다항식} \gets 유리식$  
  
- 다항식 : 분모가 상수항으로만 이루어진 꼴일 때  
- 분수식 : 분모가 상수항으로만 이루어진 꼴이 아닐 때  
  
### 유리식의 성질  
$$
\frac{B\times C}{A\times C}=\frac{B}{A}
$$  
  
### 유리식의 사칙연산  
$$
\frac{1}{x-y}+\frac{1}{x+y}=\frac{x+y+x-y}{(x-y)(x+y)}=\frac{2x}{x^2-y^2}
$$  
나눗셈 계산 : 역수 곱셈  
가분수식 계산  
$$
\frac{x^2+1}{x-1}=\frac{(x-1)(x+1)+2}{x-1}=(x+1)+\frac{2}{x-1}
$$  
다항식+진분수식 형태로 나타냄.  
  
### 부분 분수식  
$$
\frac{1}{AB}=\frac{1}{B-A}(\frac{1}{A}-\frac{1}{B})
$$
  
### 번분수식  
$$
\frac{\frac{A}{B}}{\frac{C}{D}}=\frac{AD}{BC}
$$  
  
### 비례식  
$$
\frac{a}{b}=\frac{c}{d} \iff ad=bc \iff a:b=c:d
$$  
$$
\frac{a+b}{b}=\frac{c+d}{d}
$$  
$$
\frac{a-b}{b}=\frac{c-d}{d}
$$  
$$
\frac{a+b}{a-b}=\frac{c+d}{c-d}
$$  
  
### 가비의 리  
$(b\neq0, d\neq0, b+d\neq0)$  
$$
\frac{a}{b}=\frac{c}{d}=\frac{a+c}{b+d}
$$  
  
### 유리함수  
$y=f(x), f(x)$가 유리식이면 유리함수  
$y=\frac{1}{x+3}, (x\neq-3)$  
정의역이 -3을 제외한 실수  
  
$y=\frac{ax+b}{cx+d}$ 는 $q+\frac{k}{x-p}$ 형태로 정리 가능.  
이는 $y=\frac{k}{x}$ 를 x축으로 p, y축으로 q만큼 평행이동한 것.  
  
### 유리함수의 그래프  
$y=\frac{k}{x} (k\neq0)$ 의 그래프  
정의역 {x | $x\neq0$ 인 모든 실수}  
치역 {y | $y\neq0$ 인 모든 실수}  
점근선 : x축, y축  
원점대칭, y=x 대칭, y=-x 대칭  
k>0 -> 1, 3 사분면  
k< 0 -> 2, 4 사분면  
$|k|$의 값이 크면 클수록 원점에서 멀어진다.  
일대일 대응 -> 역함수 존재(y=x에 대해 대칭이기 때문에 역함수는 원함수와 동일)  
  
$y=\frac{1}{x}$ 의 그래프
